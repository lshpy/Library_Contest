```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# 데이터 로드 (엑셀 파일 예시)
data = pd.read_excel("/Users/seunghyun/Downloads/book_analysis (4).xlsx", sheet_name=2)
# 범주형 변수 인코딩 (예: 원-핫 인코딩)
data = pd.get_dummies(data, columns=['연령', '성별'])

# 추천 여부를 기준으로 새 열 생성
threshold = data['순위'].median()  # 순위의 중간값을 기준으로 설정
data['추천'] = (data['순위'] <= threshold).astype(int)  # 순위가 중간값 이하인 경우 추천

# 특성과 레이블 분리
X = data.drop(['도서명', '대출건수', '순위', '추천'], axis=1)
y = data['추천']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 모델 생성
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# 모델 학습
clf.fit(X_train, y_train)

# 예측
y_pred = clf.predict(X_test)

# 성능 평가
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```
