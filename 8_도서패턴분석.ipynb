```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.font_manager as fm

# 데이터 불러오기
df = pd.read_excel("/Users/seunghyun/Downloads/book_analysis (6).xlsx", sheet_name=2)

# 데이터 클린징
df.dropna(inplace=True)  # 결측치 제거
df.drop_duplicates(inplace=True)  # 중복 데이터 제거

# 형식 변환
df['연령'] = df['연령'].astype(str)  # 연령을 문자열로 변환
df['성별'] = df['성별'].astype(str)

# 연령대 구분
df['연령대'] = pd.Categorical(df['연령'], categories=sorted(df['연령'].unique()), ordered=True)

# 주제분류명 정제
df['주제분류명'] = df['주제분류명'].str.strip().str.lower()  # 소문자로 변환 및 공백 제거

# 데이터 준비
age_gender_genre_count = df.groupby(['연령대', '성별', '주제분류명', '도서명'], observed=False).size().reset_index(name='대출건수')

# 기술 통계
age_gender_genre_pivot = age_gender_genre_count.pivot_table(index=['연령대', '성별', '도서명'], columns='주제분류명', values='대출건수', fill_value=0, observed=False)
print(age_gender_genre_pivot)

# 모델 구축
X = age_gender_genre_count[['연령대', '성별', '주제분류명']]
y = age_gender_genre_count['대출건수']

# 연령대와 성별을 원-핫 인코딩으로 변환
encoder = OneHotEncoder(sparse_output=False, drop='first')
X_encoded = encoder.fit_transform(X)

# 훈련 데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델 훈련
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# 모델 평가
y_pred = rf.predict(X_test)
print("정확도:", accuracy_score(y_test, y_pred))
print("분류 리포트:\n", classification_report(y_test, y_pred))

# 교차 검증
cross_val_scores = cross_val_score(rf, X_encoded, y, cv=5)
print("교차 검증 정확도 평균:", np.mean(cross_val_scores))

# 특징 중요도 분석
feature_importances = rf.feature_importances_
feature_names = encoder.get_feature_names_out()
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)

# 결과 분석 및 시각화

plt.rcParams['font.family'] ='AppleGothic.ttf'
plt.rcParams['axes.unicode_minus'] =False

# 연령대 및 성별별로 가장 선호하는 도서명 및 주제분류명
# observed=False 명시적으로 설정
top_books = age_gender_genre_count.loc[age_gender_genre_count.groupby(['연령대', '성별'], observed=False)['대출건수'].idxmax()]

print("연령대 및 성별별 가장 선호하는 도서명 및 주제분류명:")
print(top_books[['연령대', '성별', '도서명', '주제분류명', '대출건수']])

# 패턴 분석
age_gender_genre_pivot_sum = age_gender_genre_pivot.sum(axis=1).unstack()
sns.heatmap(age_gender_genre_pivot_sum, cmap='YlGnBu', annot=True, fmt='d')
plt.title('연령대 및 성별별 도서 주제 선호')
plt.show()

# 바 차트
age_gender_genre_sum = age_gender_genre_count.groupby(['연령대', '성별']).sum().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=age_gender_genre_sum, x='연령대', y='대출건수', hue='성별')
plt.title('연령대 및 성별별 도서 대출 건수')
plt.show()
```
